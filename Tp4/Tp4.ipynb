```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import square 
from math import pi

periodo = 0.2
tao = 0.05
frec = 1/periodo
duty_cicle = tao/periodo
cant_puntos = 200
cant_ciclos = 1

valores_t = np.arange(start= 0, stop= periodo*cant_ciclos, step= (periodo*cant_ciclos)/cant_puntos)

senal_cuadrada = square(2*pi*frec*valores_t,duty_cicle)

graf = plt.gcf()
graf.set_size_inches(30/2.54, 10/2.54)  # ahora en cm
plt.plot(valores_t,senal_cuadrada)
plt.title("Cuadrada")
plt.xlabel("t[s]")
plt.ylabel("V[V]")
plt.show()

cant_coef = 20
valores_n = np.arange(start= -cant_coef, stop= cant_coef + 1, step= 1)
c_n = []
for i in valores_n:
    coef = duty_cicle * np.sinc(i*duty_cicle)
    c_n.append(coef)

c_n = np.array(np.abs(c_n))

valores_f = np.arange(start=-cant_coef*frec, stop= cant_coef*frec+frec, step= frec)

graf = plt.gcf()
graf.set_size_inches(30/2.54, 10/2.54)  # ahora en cm
plt.plot(valores_f, c_n, ".", color= 'c')
plt.title("Coeficientes Serie Exponencial")
plt.xlabel("f[Hz]")
plt.ylabel("Cn")
plt.show()

arg_n = []
for i in valores_f:
    arg = -pi*i*duty_cicle
    arg_n.append(arg)

arg_n = np.array(arg_n)

graf = plt.gcf()
graf.set_size_inches(30/2.54, 10/2.54)  # ahora en cm
plt.plot(valores_f, arg_n, ".", color= 'c')
plt.title("Argumento Serie Exponencial")
plt.xlabel("f[Hz]")
plt.ylabel("Argumento")
plt.show()
```
