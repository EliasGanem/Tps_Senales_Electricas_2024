## Punto 1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import square as sq 
from math import pi

cant_ciclos = 2
cant_puntos = 1000
tao = 0.2
periodo = 2
duty_cicle = tao/periodo
frec = 1/periodo

valores_t = np.linspace(start = 0, stop = cant_ciclos*periodo, num=int(cant_ciclos*cant_puntos))

senial_cuadrada = (sq(2*pi*frec*valores_t,duty_cicle) + 1) /2

graf = plt.gcf()
graf.set_size_inches(30/2.54, 10/2.54)  # ahora en cm

plt.plot(valores_t,senial_cuadrada)
plt.title("Cuadrada")
plt.xlabel("t[s]")
plt.ylabel("V[V]")
plt.show()

cant_coef = 20
valores_n = np.arange(start=0, stop=cant_coef+1, step=1) # no son valores de frec son como valores de n
a_n = []


for i in valores_n:
    coef = 2*duty_cicle* np.sinc(i*duty_cicle) # Nota la Funcion seno cardinal tiene normalizado su argumento en pi
    a_n.append(coef)

a_n = np.array(a_n)

graf = plt.gcf()
graf.set_size_inches(30/2.54, 10/2.54)  # ahora en cm

plt.plot(valores_n,a_n,".")
plt.title("Espectro")
plt.xlabel("f")
plt.ylabel("V[V]")
plt.show()


senial = []

for i in valores_t: # para cada t discreto se hace la sumatoria infinita
    aux = 0
    for j in valores_n:
        aux = aux + 2 * duty_cicle * np.sinc(j*duty_cicle) * np.cos(2*pi*frec*j*i)
    senial.append(aux)
    
senial = np.array(senial)

graf = plt.gcf()
graf.set_size_inches(30/2.54, 10/2.54)  # ahora en cm

plt.plot(valores_t,senial)
plt.title("Cuadrada")
plt.xlabel("t[s]")
plt.ylabel("V[V]")
plt.show()


```
