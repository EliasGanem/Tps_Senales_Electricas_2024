```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
from scipy.fft import fft,fftshift,fftfreq
from scipy.signal import square,sawtooth,convolve
```

### TF de un Pulso
Esto para ver que hace lo que pienso q hace

```python
cant_puntos=200
cant_ciclos = 1
periodo = 1

amplitud = 10

t = np.arange(start= -0.5, stop= periodo*cant_ciclos-0.5, step= (periodo/cant_puntos)*cant_ciclos)

pulso_cuadrado = []
for x in t:
    if x <= 0.1 and x >=- 0.1:
        aux = amplitud
    else:
        aux = 0
    pulso_cuadrado.append(aux)
pulso_cuadrado = np.array(pulso_cuadrado)

graf = plt.gcf()
graf.set_size_inches(30/2.54, 10/2.54)  # ahora en cm
plt.plot(t, pulso_cuadrado, ".", color= 'c')
plt.title("Pulso Cuadrado")
plt.xlabel("t[s]")
plt.ylabel("x(t)")
plt.show()

fft_pulso = fft(pulso_cuadrado)
modulo = np.absolute(fft_pulso)/cant_puntos
fft_frec = fftfreq(cant_puntos,periodo/cant_puntos) ##

graf = plt.gcf()
graf.set_size_inches(30/2.54, 10/2.54)  # ahora en cm
plt.plot(fft_frec, modulo, ".", color= 'c')
plt.title("FFT")
plt.xlabel("f[hz]")
plt.xticks(np.arange(start=-100,stop=100,step=5))
plt.ylabel("Modulo")
##plt.axvline(x=0, color='black',)
plt.xticks(fontsize=8)
plt.show()

```

### Convolucion

```python
cant_puntos=500
cant_ciclos = 1
periodo = 8

amplitud_cuadrada = 1.5

t = np.arange(start= -4, stop= periodo*cant_ciclos-4, step= (periodo/cant_puntos)*cant_ciclos)

pulso_cuadrado = []
for x in t:
    if x >= -2 and x <= 1:
        aux = amplitud_cuadrada
    else:
        aux = 0
    pulso_cuadrado.append(aux)
pulso_cuadrado = np.array(pulso_cuadrado)

graf = plt.gcf()
plt.plot(t, pulso_cuadrado, color= 'g')

amplitud_triangular = 3
pulso_triangular = []
for x in t:
    if x >= 0 and x <= 2:
        aux = (3/2)*x
    else:
        aux = 0
    pulso_triangular.append(aux)
pulso_triangular = np.array(pulso_triangular)
   

plt.plot(t,pulso_triangular, color= 'b')
graf.set_size_inches(30/2.54, 10/2.54)  # ahora en cm
plt.title("Pulsos")
plt.xlabel("t[s]")
plt.axvline(x=0, color='black')
plt.ylabel("x(t)")
plt.show()

convolucion = convolve(pulso_cuadrado,pulso_triangular,mode='full') * (periodo/cant_puntos)
t_convolucion = np.linspace(start=-8,stop=8,num=len(convolucion)) ## xq de -8 a8?
graf = plt.gcf()
plt.plot(t_convolucion,convolucion, color= 'r')
graf.set_size_inches(30/2.54, 10/2.54)  # ahora en cm
plt.title("Convolucion")
plt.xlabel("t[s]")
plt.xlim(-4,4)
plt.ylabel("x*y(t)")
plt.show()
print(convolucion.size)
```
