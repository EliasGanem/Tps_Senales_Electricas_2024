# Punto 1
## a
Genere los vectores necesarios para representar alrededor de 100 ciclos de una función senoidal de 20Hz y 3v de amplitud, y otra de la misma duración con una frecuencia de 33.57Hz y 4v de amplitud.
1. Usando promedios temporales, calcule los valores eficaces de cada
señal, y de la suma y de la diferencia de cada una.
2. Compare los resultados usando la función desviación estándar (std).

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

```python
# seno_1
sin_1 = []

frec_1 = 20 #[Hz]
w_1 = 2*pi*frec_1
amp_1 = 3 #[V]

cant_ciclos = 100
frec_muestreo = 100 * frec_1

valor_x_1 = np.linspace(start=0, stop= cant_ciclos*(1/frec_1), num=int(cant_ciclos*(1/frec_1)*frec_muestreo))

for x in valor_x_1:
    valor_seno = amp_1 * np.sin(w_1*x)
    sin_1.append(valor_seno)
sin_1 = np.array(sin_1)

graf = plt.gcf()
graf.set_size_inches(25/2.54, 15/2.54)  # ahora en cm

plt.plot(valor_x_1,sin_1)
plt.title("Seno_1")
plt.xlabel("t[s]")
plt.ylabel("V[V]")
plt.show()
```

```python
#seno_2
sin_2 = []

frec_2 = 33.57 #[Hz]
w_2 = 2*pi*frec_2
amp_2 = 4 #[V]

frec_muestreo_2 = 20*frec_2

valor_x_2 = np.arange(start=0, stop=cant_ciclos*(1/frec_1), step= 1/frec_muestreo_2) # el stop es porq tiene q durar lo mismo q la señal 1

for x in valor_x_2:
    valor_seno = amp_2 * np.sin(w_2*x)
    sin_2.append(valor_seno)

sin_2 = np.array(sin_2)

graf = plt.gcf()
graf.set_size_inches(25/2.54, 15/2.54)  # ahora en cm

plt.plot(valor_x_2,sin_2)
plt.title("Seno_2")
plt.xlabel("t[s]")
plt.ylabel("V[V]")
plt.show()
```

```python
graf = plt.gcf()
graf.set_size_inches(25/2.54, 15/2.54)  # ahora en cm

plt.plot(valor_x_1,sin_1)
plt.plot(valor_x_2,sin_2)
plt.title("Seno_1 y Seno_2")
plt.xlabel("t[s]")
plt.ylabel("V[V]")
plt.show()
```

```python
valor_medio_1 = round(np.mean(sin_1),2)
valor_medio_2 = round(np.mean(sin_2),2)
print("Valor Medio seno 1 = "+str(valor_medio_1))
print("Valor Medio seno 2 = "+str(valor_medio_2))

sin_1_cuadrado = sin_1**2
sin_2_cuadrado = sin_2**2

plt.plot(valor_x_1,sin_1_cuadrado)
plt.plot(valor_x_1,sin_1)
plt.title("Seno1 y Seno 1 Cuadrado")
plt.xlabel("t[s]")
plt.ylabel("V[V]")
plt.show()

valor_cuad_medio_1 = round(np.mean(sin_1_cuadrado),2)
valor_cuad_medio_2 = round(np.mean(sin_2_cuadrado),2)
print("Valor Cuadrático Medio seno 1 = "+str(valor_cuad_medio_1))
print("Valor Cuadrático Medio seno 1 = "+str(valor_cuad_medio_2))

valor_ef_1 = round(np.sqrt(valor_cuad_medio_1),2)
valor_ef_2 = round(np.sqrt(valor_cuad_medio_2),2)

print("Valor Eficaz seno 1 = "+str(valor_ef_1))
print("Valor Eficaz seno 2 = "+str(valor_ef_2))
```

```python
 # suma = sin_1 + sin_2 # no es valido porq no poseen la misma cantidad de elementos, no estoy trabjando con una funcion, esto es un vector fila
print(len(sin_1))
print(len(sin_2))
```

Como son senos de distinta frecuencia entonces  el valor cuadratico medio de su suma o resta es la suma de sus valores cuadraticos medios

```python
valor_cuad_medio_sumayresta = valor_cuad_medio_1 + valor_cuad_medio_2
valor_ef_sumayresta = round(np.sqrt(valor_cuad_medio_sumayresta),2)
print("Valor Eficaz seno 1 + seno 2 y seno 1 - seno 2 = "+str(valor_ef_sumayresta))
```

```python
desv_est_seno_1 = round(np.std(sin_1),2)
print("Desviacion Estandar seno 1 = "+str(desv_est_seno_1))
desv_est_seno_2 = round(np.std(sin_2),2)
print("Desviacion Estandar seno 1 = "+str(desv_est_seno_2))

# notar coincide con los valores eficacez, pues la desviacion estandar es el valor eficaz de la parte alterna, y esta son señales puramente alterna
```

# Punto 2
- Genere aprox. 30 ciclos de una señal diente de sierra periódica de 15 Hz, y 200 puntos por ciclo a usando la función "mod" (modulo %).



```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

```python
cant_ciclos = 30
frec = 15 #[Hz]
puntos_ciclo = 200

valor_x = np.linspace(start=0, stop= cant_ciclos*(1/frec), num=int(cant_ciclos*puntos_ciclo))

diente_sierra = []

for ciclo in range(cant_ciclos): # range(x) genera una lista de x elementos
    for punto in range(puntos_ciclo):
        valor_senial = punto % puntos_ciclo # se usa esto porque dividir de 0-199 da un resto igual al numero, y cuando pasa a 200 da 0, luego en 201 da 1, 202 da 2, es como q reseteo la señal
        diente_sierra.append(valor_senial)
        #print(str(valor_senial))
        #print(str(diente_sierra)+"\n")
diente_sierra = np.array(diente_sierra)

#notar la amplitud es puntos_cilo-1 para el periodo
#print(valor_x)
#print(diente_sierra)

graf = plt.gcf()
graf.set_size_inches(30/2.54, 10/2.54)  # ahora en cm

plt.plot(valor_x,diente_sierra, ".")
plt.title("Diente de Sierra")
plt.xlabel("t[s]")
plt.ylabel("V[V]")
plt.show()
```

- Genere aprox. 30 ciclos de una señal cuadrada periódica de 20 Hz, y 250 puntos por ciclo, a partir de una senoidal.


```python
# Cuadrada
sin = []

frec_sin = 20
amp_sin = 5
w_sin = 2*pi*frec_sin

cant_ciclos = 30
puntos_ciclo = 250

valor_x = np.arange(start=0, stop=(1/frec_sin)*cant_ciclos, step=(1/frec_sin)/puntos_ciclo)

for x in valor_x:
    valor_seno = amp_sin * np.sin(w_sin*x)
    sin.append(valor_seno)

sin = np.array(sin)

graf = plt.gcf()
graf.set_size_inches(30/2.54, 10/2.54)  # ahora en cm
plt.plot(valor_x,sin, ".")
plt.title("Seno")
plt.xlabel("t[s]")
plt.ylabel("V[V]")
plt.show()


senial_cuadrada = []
amp_senial_cuadrada = 3

for i in sin:
    if i > 0:
        senial_cuadrada.append(amp_senial_cuadrada) 
    elif i < 0:
        senial_cuadrada.append(-amp_senial_cuadrada)
    else:
        senial_cuadrada.append(0)

senial_cuadrada = np.array(senial_cuadrada)

graf = plt.gcf()
graf.set_size_inches(30/2.54, 10/2.54)  # ahora en cm

plt.plot(valor_x,senial_cuadrada, ".")
plt.title("Cuadrada")
plt.xlabel("t[s]")
plt.ylabel("V[V]")
plt.show()
```

- Use un histograma como aproximación a la función de densidad de probabilidad de la señal del ítem 2.a. Tenga especial atención en el dimensionado de los ejes para realizar la gráfica.

```python
histograma = plt.gcf()
histograma.set_size_inches(30/2.54, 10/2.54)  # ahora en cm

plt.hist(diente_sierra, bins=200, facecolor="g", density=True, histtype="step", orientation="vertical")
plt.xlabel("f(x)")
plt.ylabel("probabilidad")
plt.show()

# density en false pone la cantidad de apariciones de f(x), en true aparece la probabilidad
# histtype da el histograma relleno
```
