Genere los vectores necesarios para representar alrededor de 100 ciclos de una función senoidal de 20Hz y 3v de amplitud, y otra de la misma duración con una frecuencia de 33.57Hz y 4v de amplitud.
1. Usando promedios temporales, calcule los valores eficaces de cada
señal, y de la suma y de la diferencia de cada una.
2. Compare los resultados usando la función desviación estándar (std).

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

```python
# seno_1
sin_1 = []

frec_1 = 20 #[Hz]
w_1 = 2*pi*frec_1
amp_1 = 3 #[V]

cant_ciclos = 100
frec_muestreo = 100 * frec_1

valor_x_1 = np.linspace(start=0, stop= cant_ciclos*(1/frec_1), num=int(cant_ciclos*(1/frec_1)*frec_muestreo))

for x in valor_x_1:
    valor_seno = amp_1 * np.sin(w_1*x)
    sin_1.append(valor_seno)
sin_1 = np.array(sin_1)

graf = plt.gcf()
graf.set_size_inches(25/2.54, 15/2.54)  # ahora en cm

plt.plot(valor_x_1,sin_1)
plt.title("Seno_1")
plt.xlabel("t[s]")
plt.ylabel("V[V]")
plt.show()
```

```python
#seno_2
sin_2 = []

frec_2 = 33.57 #[Hz]
w_2 = 2*pi*frec_2
amp_2 = 4 #[V]

frec_muestreo_2 = 20*frec_2

valor_x_2 = np.arange(start=0, stop=cant_ciclos*(1/frec_1), step= 1/frec_muestreo_2) # el stop es porq tiene q durar lo mismo q la señal 1

for x in valor_x_2:
    valor_seno = amp_2 * np.sin(w_2*x)
    sin_2.append(valor_seno)

sin_2 = np.array(sin_2)

graf = plt.gcf()
graf.set_size_inches(25/2.54, 15/2.54)  # ahora en cm

plt.plot(valor_x_2,sin_2)
plt.title("Seno_2")
plt.xlabel("t[s]")
plt.ylabel("V[V]")
plt.show()
```

```python
graf = plt.gcf()
graf.set_size_inches(25/2.54, 15/2.54)  # ahora en cm

plt.plot(valor_x_1,sin_1)
plt.plot(valor_x_2,sin_2)
plt.title("Seno_1 y Seno_2")
plt.xlabel("t[s]")
plt.ylabel("V[V]")
plt.show()
```

```python
valor_medio_1 = round(np.mean(sin_1),2)
valor_medio_2 = round(np.mean(sin_2),2)
print("Valor Medio seno 1 = "+str(valor_medio_1))
print("Valor Medio seno 2 = "+str(valor_medio_2))

sin_1_cuadrado = sin_1**2
sin_2_cuadrado = sin_2**2

plt.plot(valor_x_1,sin_1_cuadrado)
plt.plot(valor_x_1,sin_1)
plt.title("Seno1 y Seno 1 Cuadrado")
plt.xlabel("t[s]")
plt.ylabel("V[V]")
plt.show()

valor_cuad_medio_1 = round(np.mean(sin_1_cuadrado),2)
valor_cuad_medio_2 = round(np.mean(sin_2_cuadrado),2)
print("Valor Cuadrático Medio seno 1 = "+str(valor_cuad_medio_1))
print("Valor Cuadrático Medio seno 1 = "+str(valor_cuad_medio_2))

valor_ef_1 = round(np.sqrt(valor_cuad_medio_1),2)
valor_ef_2 = round(np.sqrt(valor_cuad_medio_2),2)

print("Valor Eficaz seno 1 = "+str(valor_ef_1))
print("Valor Eficaz seno 2 = "+str(valor_ef_2))
```

```python
 # suma = sin_1 + sin_2 # no es valido porq no poseen la misma cantidad de elementos, no estoy trabjando con una funcion, esto es un vector fila
print(len(sin_1))
print(len(sin_2))
```

Como son senos de distinta frecuencia entonces  el valor cuadratico medio de su suma o resta es la suma de sus valores cuadraticos medios

```python
valor_cuad_medio_sumayresta = valor_cuad_medio_1 + valor_cuad_medio_2
valor_ef_sumayresta = round(np.sqrt(valor_cuad_medio_sumayresta),2)
print("Valor Eficaz seno 1 + seno 2 y seno 1 - seno 2 = "+str(valor_ef_sumayresta))
```

```python
desv_est_seno_1 = round(np.std(sin_1),2)
print("Desviacion Estandar seno 1 = "+str(desv_est_seno_1))
desv_est_seno_2 = round(np.std(sin_2),2)
print("Desviacion Estandar seno 1 = "+str(desv_est_seno_2))

# notar coincide con los valores eficacez, pues la desviacion estandar es el valor eficaz de la parte alterna, y esta son señales puramente alterna
```
